```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist() # list of movies watched by the user
movies_watched[0:20] 
```

```python
len(movies_watched)              
```

```python


movies_watched_df = user_movie_df[movies_watched] # Those who watched at least one of the movies the user watched
movies_watched_df.head()

```

```python


movies_watched_df=movies_watched_df[~movies_watched_df[movies_watched[0]].isnull()]
```

```python
user_movie_count = movies_watched_df.T.notnull().sum().reset_index() #Discarding null values and making some index corrections
user_movie_count.columns = ["userId", "movie_count"]
user_movie_count
```

```python


k=len(movies_watched)*0.60   # this variable is used for detect the similarity, it is depended number of movies watched.
k # the users that watched the movies below k are excluded 

```

```python
users_same_movies=user_movie_count[user_movie_count["movie_count"] > k]["userId"]
users_same_movies
```

```python
final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movies)],
                      random_user_df[movies_watched]])
final_df.head()
```

```python
corr_df=final_df.T.corr().unstack().sort_values().drop_duplicates() #Correlation for UserA X UserB
corr_df
```

```python


corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df

```

```python


corr_df.index.names = ['user_id_1', 'user_id_2']

```

```python


corr_df = corr_df.reset_index()
corr_df

```

```python
top_users=corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.50)][["user_id_2","corr"]].reset_index(drop=True)
```

```python
top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
top_users=top_users[top_users["userId"]!=random_user]
top_users
```

```python
top_users_ratings = top_users.merge(df_R[["userId", "movieId", "rating"]], how='inner')
top_users_ratings
```

```python


top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']


top_users_ratings.head()

```

```python


recommendation_df=top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})

```

```python
recommendation_df = recommendation_df.reset_index()
```
